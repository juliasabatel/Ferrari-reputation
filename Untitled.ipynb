```python
import requests
from bs4 import BeautifulSoup

# URL de la página a scrapear
url = 'https://www.marca.com/motor/carlos-sainz-jr.html'

# Realizar la solicitud GET para obtener el contenido de la página
response = requests.get(url)

# Comprobar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Analizar el contenido HTML de la página con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscar todos los enlaces que contienen los artículos (modifica la clase si es necesario)
    articulos = soup.find_all('a', class_='article-link')

    # Mostrar los títulos y enlaces de los artículos
    for articulo in articulos:
        titulo = articulo.get_text(strip=True)
        enlace = articulo.get('href')
        print(f'Título: {titulo}')
        print(f'Enlace: {enlace}')
        print('---')
else:
    print(f"Error al obtener la página. Código de estado: {response.status_code}")
```
